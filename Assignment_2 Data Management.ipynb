```python
#6.Calculate the sum for variable ‘Sum_Assured’ by ‘Region’ variable.

Sum_Assured_byRegion = Premium.groupby('REGION')['Sum_Assured'].sum().reset_index()

print(Sum_Assured_byRegion)

##groupby  function is used
```

```python
#7.Create a subset of policies of Asia Standard Plan with Sum_Assured < = 50,000.
#Keep variables Policy_No, Zone_name, Plan and Sum_Assured in the subset data.
Subset_of_Policies=Premium.loc[(Premium.Plan=='Asia Standard Plan') &
(Premium.Sum_Assured<=50000), ['POLICY_NO','ZONE_NAME', 'Plan','Sum_Assured']]
print(Subset_of_Policies)


```

```python
#8. Export the subsetted data into an xlsx file.
Subset_of_Policies.to_excel('C:/Users/datam/Desktop/CSVs/Subset_of_Policies.xlsx', index=False)
print("Exported the subsetted data into an xlsx file")
```
