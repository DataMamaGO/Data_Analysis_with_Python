```python
#2.For each zone, obtain the mean Premium and plot a bar chart showing the mean Premium over zone. 
#(Use any color from a palette from R( Color Brewer)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
Premium = pd.read_csv('C:/Users/datam/Desktop/CSVs/Premiums.csv')
Premium_mean= Premium.groupby('ZONE_NAME')['Premium'].sum()
Premium_mean
print(Premium_mean)
plt.figure(); Premium_mean.plot.bar(title=' BAR CHART (Premium mean by Zone group)', color='purple'); plt.xlabel('Zone group');
plt.ylabel('Premium mean by Zone')
```

```python
# Bar Chart in Horizontal orientation
plt.figure();Premium_mean.plot.barh(title=' BAR CHART Premium mean by Zone HORIZONTAL', color='lightgreen'); plt.xlabel('Zone group');
plt.ylabel('Premium mean')
```

```python
#3.Obtain a stacked bar chart for all the Zones over Sub plans by the Premium amount. 


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Premium = pd.read_csv('C:/Users/datam/Desktop/CSVs/Premiums.csv')

# Create a pivot table
Premium_amount = pd.pivot_table(Premium, index='ZONE_NAME', columns='Sub_Plan', values='Premium', aggfunc='sum')
print(Premium_amount)
# Plot the data
Premium_amount.plot(kind='bar', stacked=True, title='STACKED BAR CHART')
plt.xlabel('Zone name')
plt.ylabel('Premium amount')
plt.show()

```

```python
#4.Obtain a heat map of Plan and Zone with respective average Premium.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Premium = pd.read_csv('C:/Users/datam/Desktop/CSVs/Premiums.csv')

Premium_amount = pd.pivot_table(Premium, values='Premium', index='Plan', columns='ZONE_NAME', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(Premium_amount, annot=True, cmap='twilight', fmt='.2f', linewidths=.5)
plt.title('Heat Map')
plt.show()
```

```python
#5.Obtain a pie chart using ggplot2 for Premium amount across different sub plans.
#(Use any palette from R (Color Brewer)  
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Premium = pd.read_csv('C:/Users/datam/Desktop/CSVs/Premiums.csv')
Premium_amount = pd.pivot_table(Premium, index='ZONE_NAME', columns='Sub_Plan', values='Premium', aggfunc='sum')
premium_sums = Premium_amount.sum()

# Plot the pie chart
premium_sums.plot.pie(
    labels=premium_sums.index,
    title="PIE CHART WITH PERCENTAGE",
    colormap='brg',
    autopct='%1.0f%%'
)

plt.ylabel('')  
plt.show()
```
