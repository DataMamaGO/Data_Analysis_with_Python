```python
#2.Obtain the Mode for the count of policies available across each Zone.

import pandas as pd
Premium = pd.read_csv('C:/Users/datam/Desktop/CSVs/Premiums.csv')
zone_counts = Premium.groupby('ZONE_NAME').size()
mode_count = zone_counts.mode()
print(zone_counts)
print(mode_count)
```

```python
#3.Obtain box-whisker plots for Vintage period. 
import pandas as pd
import matplotlib.pyplot as plt
Premium = pd.read_csv('C:/Users/datam/Desktop/CSVs/Premiums.csv')

df = pd.DataFrame(Premium)
plt.figure(figsize=(10, 6))
plt.boxplot(df['Vintage_Period'])
plt.title('Box-Whisker Plot for Vintage_Period')
plt.ylabel('Vintage_Period')

plt.show()

```

```python
#4.Detect outliers if present. Hint: use Boxplot() function of ‘car’ Package
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
Premium.boxplot(column='Vintage_Period', by='ZONE_NAME')
plt.title('Boxplot of Vintage_Period by Zone')
plt.suptitle('')
plt.xlabel('Zone Name')
plt.ylabel('Vintage_Period')
plt.show()
```

```python
#5.Find skewness and kurtosis of Premium amount by Zone.
import pandas as pd
import scipy.stats as stats
Premium = pd.read_csv('C:/Users/datam/Desktop/CSVs/Premiums.csv')
Premium=pd.DataFrame(Premium)
skewness = Premium.groupby('ZONE_NAME')['Premium'].apply(stats.skew)
kurtosis = Premium.groupby('ZONE_NAME')['Premium'].apply(stats.kurtosis)
print("Skewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)
```

```python
#6.Draw a scatter plot of Premium and Vintage period. 
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(Premium['Vintage_Period'], Premium['Premium'], c='blue', marker='o')
plt.title('Scatter Plot of Premium and Vintage Period')
plt.xlabel('Vintage Period')
plt.ylabel('Premium')
plt.grid(True)
plt.show()
```

```python
#7.Find the correlation coefficient between Premium and Vintage period and interpret the value.
import pandas as pd
import numpy as np
correlation = np.corrcoef(Premium['Premium'], Premium['Vintage_Period'])[0, 1]
print("Correlation coefficient between Premium and Vintage period:", correlation)
```
